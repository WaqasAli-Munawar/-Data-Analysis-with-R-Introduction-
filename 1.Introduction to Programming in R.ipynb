In the digital age, data has become very important. Its analysis permits discovering hidden information that contributes, for example, to understanding a process or improving a business. This has made data science a popular discipline in recent years.

**What is data science** Simply put, Data science is an interdisciplinary field that aims to extract knowledge from data. This extraction involves several stages of data exploration, manipulation, and analysis.

**Why people should learn data science** Data science is one of the most exciting and fastest-growing fields out there. Companies such as Google, Facebook, and Amazon have demonstrated how using data they can optimize their businesses, anticipate and predict customer's reactions and make life easier for people through powerful web search engines and smartphone personal assistants. It's clear that the jobs market is being restructured, and the demand for data science skills is only increasing.

**When to use/learn Data Analysis** Data analysis requires two elements: data (at the core of the process) and questions (objectives). Learn data science today: If we will have those two elements in the future.

**How to learn data science** There are several tools for data science. We'll be using R. R is a powerful and comprehensive programming language for data analysis with an intuitive syntax.



Data scientists extract information from data and use it to create valuable predictions, visualizations, and technologies. To learn from data, we often need to perform billions of computations over large datasets. We do this with the aid of computers, which need to be given a set of instructions. We refer to writing these instructions as **programming**.

There are a variety of **computer languages** that we can use for programming. We'll focus on learning `R`, a language that offers excellent support for data science work.

The instructions we write using a computer language are called **code**. The code we write to instruct the computer to perform a task is referred to as a **program**.

We will start our `R` learning journey by discovering the **literal values**. A literal value is any part of a statement or expression that is to be used exactly as it's meant to be interpreted.

For instance, the numbers 125 and 3, are literal values.

Every literal has a **type**. It exists in R five basic data types: `Logical`, `Integer`, `Numeric`, `Character`, and `Complex`. We're not covering the Complex data type in our path because it is an advanced data type that won't be typically useful for us here.

There are two `logical` literals — `"true"` and `"false"`. In R, we write respectively `TRUE` and `FALSE` or simply `T` and `F`. All the letters are in UPPERCASE. This is important because R is **case-sensitive**, i.e., it makes a difference between upper and lower cases.

Typically, in programming languages, the logical literals represent two values: `1` (for true) and `0` (for false). Similarly, `R` also counts the `TRUE` literal as `1` and the `FALSE` literal as `0`.

Logical literals are typically used in R for testing. For example, to compare two values (greater, less, or equal). The logical value is the result of the test.

There are also two types of numbers: integers (Integer) and decimal/real (Numeric) numbers.

* `125`, `3`, `122` are integer numbers.
* `0.50`, `3.14`, `0.00009` are decimal numbers.

By default, R considers any number (integer and decimal) as numeric literals. What if we want to represent a "real" integer then?

To force R to consider numbers as integer literals (without decimal part), the number must be accompanied by the letter `L`. Note that, ONLY the capitalized 'L' letter is admitted.

`125L`

Sometimes we want to use words and characters as they are without any interpretation from R. These are called **character** literals.

Character literals are represented by wrapping them in double quotation marks (`"`) or single quotation marks (`'`). Quotes or apostrophes allow distinguishing character values to other literal values (Integer, Logical, and Numeric).

Note that if we use "FALSE" (with quotes) R doesn't consider it as a logical literal but as a character literal. Similarly, `"12.4"` is not deemed as a numeric literal but as a character literal as well.

Programming consists of providing the computer with a **set of instructions** to perform some tasks. This set of instructions is called a **program**.

In programming terminology, a program is called **code** and each instruction (line) in that code is called a **line of code**.

R allows only one instruction per line of code, e.g., if we enter two logical literals on the same line, we will get an error. This justifies why at each instruction, we insist on having one answer per line.
